```python
import torch
from torch.autograd import Variable
import torch.nn.functional as F

x_data = Variable(torch.Tensor([[1.0], [2.0], [3.0], [4.0]]))
y_data = Variable(torch.Tensor([[0.], [0.], [1.], [1.]]))
```

```python
class Model(torch.nn.Module):
    def __init__(self):
        #inisiasi constructor
        super(Model, self).__init__()
        self.linear = torch.nn.Linear(1,1)
        
    def forward(self,x):
#         y_pred = self.linear(x)
        y_pred = F.sigmoid(self.linear(x))
        return y_pred
    
#model
model = Model()

#construct loss function and an optimizer
#in the SGD constructor will contain the learnable parameters of the two
#nn.Linear modules which are members of the model
criterion = torch.nn.BCELoss(size_average=True)
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)

#Training loop
for epoch in range (1000):
    #forward pass
    y_pred = model(x_data)
    
    #compute and print loss
    loss = criterion(y_pred,y_data)
    print(epoch, loss.item())
    
    #zero gradient and perform backward and update weigth
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
#After training
hour_var = Variable(torch.Tensor([[1.0]]))
# y_pred = model(hour_var)
print("predict (1 hour)",1.0, model(hour_var).item() > 5.0)

hour_var = Variable(torch.Tensor([[7.0]]))
# y_pred = model(hour_var)
print("predict (7 hour)",7.0, model(hour_var).item()>5.0)
```
