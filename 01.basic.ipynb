```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_data = [1.0,2.0,3.0]
y_data = [2.0,4.0,6.0]

#model forward pass
# y = wx
def forward(x):
    return x*w

#loss function
#error (y_pred-y_data)^2
def loss_function(x,y):
    y_pred = forward(x)
    return (y_pred - y)**2

w_list =[]
mse_list = []
l_sum = 0

for w in np.arange(0.0,4.1,0.1):
    print("w = ", w)
    for x_val,y_val in zip(x_data, y_data):
        y_pred_val = forward(x_val)
        l = loss_function(x_val,y_val)
        l_sum += l
        print("x_val: ",x_val," y_val: ",y_val," y_pred: ",y_pred_val, " loss: ",l)
        
    mse = l_sum/3
    print("MSE: ", mse)
    w_list.append(w)
    mse_list.append(mse)

plt.plot(w_list,mse_list)
plt.ylabel("loss")
plt.xlabel("weight")
plt.show()
```
